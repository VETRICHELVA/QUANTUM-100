Cell 1 — Imports

```python
import numpy as np
```

Cell 2 — Define ALL Basic Gates

```python
X = np.array([[0,1],[1,0]],dtype=complex)
H = (1/np.sqrt(2))*(np.array([[1,1],[1,-1]],dtype=complex))
I = np.array([[1,0],[0,1]],dtype=complex)
Z = np.array([[1,0],[0,-1]], dtype=complex)
Y = np.array([[0,-1j],[1j,0]],dtype=complex)
S = np.array([[1,0],[0,1j]],dtype=complex)
T = np.array([[1,0],[0,np.exp(1j*np.pi/4)]],dtype = complex)
```

Cell 3 — Improved Qubit Class

```python
class Qubit:
    def __init__(self, alpha=1, beta=0):
        self.state = np.array([alpha, beta], dtype=complex)
        self.normalize()

    def normalize(self):
        self.state = self.state / np.linalg.norm(self.state)

    def probabilities(self):
        return abs(self.state[0])**2, abs(self.state[1])**2

    def apply_gate(self, gate):
        self.state = gate @ self.state
        self.normalize()

    def measure(self):
        p0, p1 = self.probabilities()
        result = np.random.choice([0,1], p=[p0,p1])
        self.state = np.array([1,0]) if result==0 else np.array([0,1])
        return result

    def show(self):
        p0, p1 = self.probabilities()
        print("\nState vector:", np.round(self.state,3))
        print(f"P(0)={p0:.3f}, P(1)={p1:.3f}")
```

Cell 4 — Gate Dictionary

```python
GATES = {
    "I": I,
    "X": X,
    "Y": Y,
    "Z": Z,
    "H": H,
    "S": S,
    "T": T
}
```

Cell 5 — Apply Multiple Gates

```python
def run_circuit(sequence):
    q = Qubit()
    print("Initial |0⟩")
    q.show()

    for g in sequence.split():
        if g.upper() in GATES:
            q.apply_gate(GATES[g.upper()])
            print(f"\nAfter {g.upper()} gate:")
            q.show()
        else:
            print(f"Unknown gate: {g}")

    print("\nMeasurement:", q.measure())
    q.show()
```
